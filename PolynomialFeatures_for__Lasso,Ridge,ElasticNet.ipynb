```python
#import libraraies
# read file
import pandas as pd
import numpy as np
house_df = pd.read_csv('house.csv')
house_df
```

```python
#count num values
house_df.count()
```

```python
#drop first column
house_df=house_df.drop(house_df.columns[0],axis=1)
house_df
```

```python
# normalize it
from sklearn.preprocessing import MinMaxScaler
house_df = MinMaxScaler().fit_transform(house_df)
house_df
```

```python
#show it in one value instead of range
house_df = pd.DataFrame(house_df)
house_df
```

```python
#determine input  for the model
x=house_df.iloc[:,:-1]
x
```

```python
#determine output for the model
y=house_df.iloc[:,-1]
y
```

```python
#split the data set to training : test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python

#do linear regression
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression().fit(x_train,y_train)
print('LinearRegression tarining accuracy : '+ str(lin_reg.score(x_train,y_train)))
print('LinearRegression testing accuracy : '+ str(lin_reg.score(x_test,y_test)))
```

```python

#find error

from sklearn.metrics import mean_squared_error
y_pred=lin_reg.predict(x_train)
print(' tarining error : '+ str(mean_squared_error(y_train,y_pred)))
from sklearn.metrics import mean_squared_error
y_preds=lin_reg.predict(x_test)
print(' testing accuracy : '+ str(mean_squared_error(y_test,y_preds)))
```

```python
#applay PolynomialFeatures then do linear regression
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
x_poly.shape

x_train,x_test,y_train,y_test=train_test_split(x_poly,y,test_size=0.3,random_state=42)
lin_reg=LinearRegression().fit(x_train,y_train)
print('LinearRegression tarining accuracy after PolynomialFeatures : '+ str(lin_reg.score(x_train,y_train)))
print('LinearRegression testing accuracy after PolynomialFeatures : '+ str(lin_reg.score(x_test,y_test)))

#applay lassomodel, ridgemodel, elasticmodel
from sklearn.linear_model import Lasso,Ridge,ElasticNet

lasso=Lasso().fit(x_train, y_train)
print('Lasso tarining accuracy : '+ str(lasso.score(x_train,y_train)))
print('Lasso testing accuracy : '+ str(lasso.score(x_test,y_test)))

ridge=Ridge().fit(x_train, y_train)
print('Ridge tarining accuracy : '+ str(ridge.score(x_train,y_train)))
print('Ridge testing accuracy : '+ str(ridge.score(x_test,y_test)))

elastic=ElasticNet().fit(x_train, y_train)
print('ElasticNet tarining accuracy : '+ str(elastic.score(x_train,y_train)))
print('ElasticNet testing accuracy : '+ str(elastic.score(x_test,y_test)))
```
